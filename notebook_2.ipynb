# Import the data set

```python
#import the libraries

%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#to find out where my working directory is
import os
print(os.getcwd())

# convert scientific notation to decimals
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
#import data set

df = pd.read_csv('indicators.csv')
```

# First 1-variable analysis
Lets perform a one-variable analysis on life expectancy from 1960-2013

```python
#Here I am defining a new dataframe and passing the information I want into it from the original source.

df_life_expectancy = df[df['IndicatorCode'].str.contains('SP.DYN.LE00.IN')]

df_life_expectancy_canada = df_life_expectancy[df_life_expectancy['CountryName'].str.contains('Canada')]  

df_life_expectancy_canada["Value"].mean()
```

# Second 1-variable analysis
Lets look at metric tons of C02 emissions per capita in india

```python
#Here I am doing the same sort of aggregation as above, except with 

df_co2 = df[df['IndicatorCode'].str.contains('EN.ATM.CO2E.PC')]

df_co2_india = df_co2[df_co2['CountryName'].str.contains('India')]  

df_co2_india[['Year', 'Value']]
```

# Third 1-variable analysis
Lets take a look at proportion of woman holding seats in parliament in Russia

```python
df_woman_seats = df[df['IndicatorCode'].str.contains('SG.GEN.PARL.ZS')]

df_woman_seats_russia = df_woman_seats[df_woman_seats['CountryName'].str.contains('Russia')]

df_woman_seats_russia
```

# First multi-variable analysis
I will look at GDP and C02 emissions. Lets focus on the united states and see if there's a correlation.
Looking at the two tables GDP rises very much unlike CO2 emissions which rise, and then fall again over the same timeline. This could be due to the fact that the united states moved much of its manufacturing overseas.

```python
df_gdp = df[df['IndicatorCode'].str.contains('NY.GDP.PCAP.CD')]
df_gdp_us = df_gdp[df_gdp['CountryName'].str.contains('United States')]
df_gdp_us[['Year', 'Value']]
```

```python
df_co2 = df[df['IndicatorCode'].str.contains('EN.ATM.CO2E.PC')]
df_co2_us = df_co2[df_co2['CountryName'].str.contains('United States')]
df_co2_us[['Year', 'Value']]
```

# Second multi-variable analysis
Lets do the same variables as last time except with China.

```python
df_gdp = df[df['IndicatorCode'].str.contains('NY.GDP.PCAP.CD')]
df_gdp_china = df_gdp[df_gdp['CountryName'].str.contains('China')]
df_gdp_china[['Year', 'Value']]
```

```python
df_co2 = df[df['IndicatorCode'].str.contains('EN.ATM.CO2E.PC')]
df_co2_china = df_co2[df_co2['CountryName'].str.contains('China')]
df_co2_china[['Year', 'Value']]
```

# Third multi-variable analysis
I will now take a look at fertility rate (births per 1000 women) and urban population % in China

```python
df_fert = df[df['IndicatorCode'].str.contains('SP.DYN.TFRT.IN')]
df_fert_china = df_fert[df_fert['CountryName'].str.contains('China')]
df_fert_china[['Year', 'Value']]
```

```python
df_urban_pop = df[df['IndicatorCode'].str.contains('SP.URB.TOTL.IN.ZS')]
df_urban_pop_china = df_urban_pop[df_urban_pop['CountryName'].str.contains('China')]
df_urban_pop_china[['Year', 'Value']]
```

# Descriptive analysis 1
Q: How many women hold seats in parliament in russia?

A: With reference to 'Third 1-variable analysis' it is currently at 13% (2014) of the total seats held. Looking at the graph we can see a step decline up until 2006. The business value is this stat could be used by a political party to promote change in government.


```python
plt.figure(figsize=(12,6))
df_group = df_woman_seats_russia.groupby(by='Year')['Value'].sum()
plt.plot(df_group.index, df_group.values, '-')
plt.show()
```

# Descriptive analysis 2
Q: What is the co2 emissions per capita in India?

A: As we can see by the graph there is a very clear trend that the CO2 emissions per capita is rising. Especially from the 2000's onward it seems the slope has gotten steeper. The possible business value from this is that it could be used by NGO's as statistical backing to fight climate change.

```python
plt.figure(figsize=(10,5))
df_group = df_co2_india.groupby(by='Year')['Value'].sum()
plt.plot(df_group.index, df_group.values, '-')
plt.show()
```

# Descriptive analysis 3
Q: What is the average life expectancy in Canada?

A: Since the 1960's, the life expectancy of canada has risen by about 10 years. The average was about 70 in the 60's and now is hovering just above 80 years old.

```python
plt.figure(figsize=(12,6))
df_group = df_life_expectancy_canada.groupby(by='Year')['Value'].sum()
plt.plot(df_group.index, df_group.values, '-')
plt.show()
```
